```python
from bs4 import BeautifulSoup
import requests
from pprint import pprint
import pandas as pd
```

```python
# load the data
data = pd.read_csv("sp500.csv") 
```

```python
# for data to be jsonifyed
stock_dictionary = {}
for stock_name in list(data.Symbol):
    
    url = "https://finance.yahoo.com/quote/"+stock_name
    request = requests.get(url)
    soup4 = BeautifulSoup(request.text, 'html.parser')
    result = soup4.find_all("table")
    if result != []:
        
        try:
            previous_close_number = float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[0])\
                                              .split("<")[1].split(">")[1])
        except:

            previous_close_number = "N/A"
            





        try:
            open_number = float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[1])\
                                .split("<")[1].split(">")[1])
        except:
            open_number = "N/A"




        try:
            bid_number = str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[2])\
                .split("<")[1].split(">")[1]
        except:
            bid_number = "N/A"
  

        try:
            ask_number = str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[3])\
                .split("<")[1].split(">")[1]
        except:
            ask_number = "N/A"


        try:
            day_range_str = str(result[0].find_all(attrs={'class','Ta(end)'})[4])\
                .split("<")[1].split(">")[1]
        except: 
            day_range_str = "N/A"


        try:
            fifty_two_week_range_str = str(result[0].find_all(attrs=\
                {'class','Ta(end)'})[5]).split("<")[1].split(">")[1]
        except:
            fifty_two_week_range_str="N/A"

        # the purpose of the if statements is to take the comma out of the volume (of the stock) number
        # because the comma prevents the number from becoming a float
        # it is first checking if there is one comma in the string then 
        # checking if there is two string in the volume
        # if there is one comma in the string the first if statement will return as true
        # if there are two commas in the string then the second if statement will return as true

        try:
            volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[0]+
                str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[1])
        except:
            volume_number = "N/A"

          #volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[0]+
    #         str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[1]+
    #         str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[2])




        #the same thing we did for the volume we are going to do for the average volume
        # because the comma in the string will prevent it from becoming a float


    #         average_volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[0]+\
    #         str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[1])

        try:
            average_volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[0]+\
                str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[1]+\
                str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[2])
        except:
            average_volume_number = "N/A"




        try:
            market_cap_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[0])\
                                    .split("<")[1].split(">")[1].split("B")[0])
        except:
            market_cap_number = "N/A"




        try:
            beta_3y_monthly_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[1])\
                                           .split("<")[1].split(">")[1])
        except:
            beta_3y_monthly_number = "N/A"



        try:
            pe_ratio_number = float(str(result[1]\
                 .find_all(attrs={'class','Trsdu(0.3s)'})[2])\
                  .split("<")[1].split(">")[1])
        except:
            pe_ratio_number = "N/A"




        try:
            eps_ttm_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[3])
                .split("<")[1].split(">")[1])
        except:
            eps_ttm_number = "N/A"
            


        try:
            earning_date = str(result[1].find_all\
                (attrs={'class',"Ta(end) Fw(600) Lh(14px)"})[4])\
                .split("<")[1].split(">")[1]
        except:
            earning_date = "N/A"




        try:
            forward_dividend_number = float(str(result[1].find_all\
                (attrs={'class',"Ta(end) Fw(600) Lh(14px)"})[5])\
                .split("<")[1].split(">")[1]\
                .split(" ")[0])
        except:
            forward_dividend_number = "N/A"






#         try:
#             forward_yeild_number = float(str(result[1].find_all\
#                 (attrs={'class',"Ta(end) Fw(600) Lh(14px)"})[5])\
#                 .split("<")[1].split(">")[1]\
#                 .split(" ")[1].split("%")[0]\
#                 .split("(")[1])
#         except ValueError:
#             forward_yeild_number = "N/A"



        try:
            ex_dividend_date = str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[4])\
                .split("<")[1].split(">")[1]
        except:
            ex_dividend_date="N/A"



        try:
            one_year_target_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[5])\
                .split("<")[1].split(">")[1])
        except: 
            one_year_target_number = "N/A"

        stock_dictionary.update({stock_name:{
        "stock_name":stock_name,
        "previous_close":previous_close_number,
        "open":open_number,
        "bid":bid_number,
        "ask":ask_number,
        "day_range":day_range_str,
        "52week_range": fifty_two_week_range_str,
        "volume": volume_number,
        "avg_volume":average_volume_number,
        "market cap": market_cap_number,
        "beta": beta_3y_monthly_number,
        "pe_ratio": pe_ratio_number,
        "eps":eps_ttm_number,
        "earnings_date":earning_date,
        "forward_div":forward_dividend_number,
        #"forward_yeild": forward_yeild_number,
        "ex_div_date":ex_dividend_date,
        "1year_target_est":one_year_target_number

        }})
```

```python
# for data to be put into a pandas dataframe

stock_name_list=[]
previous_close_list=[]
open_list = []
bid_list=[]
ask_list = []
day_range_list =[]
week_52_range_list=[]
volume_list=[]
avg_volume_list=[]
market_cap_list=[]
beta_list=[]
pe_ratio_list=[]
eps_list=[]
earnings_date_list=[]
forward_div_list=[]
forward_yeild_list=[]
ex_div_date_list=[]
one_year_target_est_list=[]
for stock_name in list(data.Symbol):
    
    url = "https://finance.yahoo.com/quote/"+stock_name
    request = requests.get(url)
    soup4 = BeautifulSoup(request.text, 'html.parser')
    result = soup4.find_all("table")
    if result != []:
        
        try:
            previous_close_number = float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[0])\
                                              .split("<")[1].split(">")[1])
        except:

            previous_close_number = "N/A"





        try:
            open_number = float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[1])\
                                .split("<")[1].split(">")[1])
        except:
            open_number = "N/A"




        try:
            bid_number = str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[2])\
                .split("<")[1].split(">")[1]
        except:
            bid_number = "N/A"
  

        try:
            ask_number = str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[3])\
                .split("<")[1].split(">")[1]
        except:
            ask_number = "N/A"


        try:
            day_range_str = str(result[0].find_all(attrs={'class','Ta(end)'})[4])\
                .split("<")[1].split(">")[1]
        except: 
            day_range_str = "N/A"



        try:
            fifty_two_week_range_str = str(result[0].find_all(attrs=\
            {'class','Ta(end)'})[5]).split("<")[1].split(">")[1]
        except:
            fifty_two_week_range_str="N/A"

        # the purpose of the if statements is to take the comma out of the volume (of the stock) number
        # because the comma prevents the number from becoming a float
        # it is first checking if there is one comma in the string then 
        # checking if there is two string in the volume
        # if there is one comma in the string the first if statement will return as true
        # if there are two commas in the string then the second if statement will return as true

        try:
            volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[0]+
                str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[1])
        except:
            volume_number = "N/A"

          #volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[0]+
    #         str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[1]+
    #         str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[2])




        #the same thing we did for the volume we are going to do for the average volume
        # because the comma in the string will prevent it from becoming a float


    #         average_volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[0]+\
    #         str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[1])

        try:
            average_volume_number =float(str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[4]).split("<")[1].split(">")[1].split(",")[0]+\
                str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[1]+\
                str(result[0].find_all(attrs={'class','Trsdu(0.3s)'})[5]).split("<")[1].split(">")[1].split(",")[2])
        except:
            average_volume_number = "N/A"




        try:
            market_cap_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[0])\
                                    .split("<")[1].split(">")[1].split("B")[0])
        except:
            market_cap_number = "N/A"




        try:
            beta_3y_monthly_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[1])\
                                           .split("<")[1].split(">")[1])
        except:
            beta_3y_monthly_number = "N/A"



        try:
            pe_ratio_number = float(str(result[1]\
                 .find_all(attrs={'class','Trsdu(0.3s)'})[2])\
                  .split("<")[1].split(">")[1])
        except:
            pe_ratio_number = "N/A"




        try:
            eps_ttm_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[3])
                .split("<")[1].split(">")[1])
        except:
            eps_ttm_number = "N/A"
           


        try:
            earning_date = str(result[1].find_all\
                (attrs={'class',"Ta(end) Fw(600) Lh(14px)"})[4])\
                .split("<")[1].split(">")[1]
        except:
            earning_date = "N/A"




        try:
            forward_dividend_number = float(str(result[1].find_all\
                (attrs={'class',"Ta(end) Fw(600) Lh(14px)"})[5])\
                .split("<")[1].split(">")[1]\
                .split(" ")[0])
        except:
            forward_dividend_number = "N/A"






        try:
            forward_yeild_number = float(str(result[1].find_all\
                (attrs={'class',"Ta(end) Fw(600) Lh(14px)"})[5])\
                .split("<")[1].split(">")[1]\
                .split(" ")[1].split("%")[0]\
                .split("(")[1])
        except:
            forward_yeild_number = "N/A"





        try:
            ex_dividend_date = str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[4])\
            .split("<")[1].split(">")[1]
        except:
            ex_dividend_date="N/A"



        try:
            one_year_target_number = float(str(result[1].find_all(attrs={'class','Trsdu(0.3s)'})[5])\
                .split("<")[1].split(">")[1])
        except:
            one_year_target_number = "N/A"

        

        stock_name_list.append(stock_name)
        previous_close_list.append(previous_close_number)
        open_list.append(open_number)
        bid_list.append(bid_number)
        ask_list.append(ask_number)
        day_range_list.append(day_range_str)
        week_52_range_list.append(fifty_two_week_range_str)
        volume_list.append(volume_number)
        avg_volume_list.append(average_volume_number)
        market_cap_list.append(market_cap_number)
        beta_list.append(beta_3y_monthly_number)
        pe_ratio_list.append(pe_ratio_number)
        eps_list.append(eps_ttm_number)
        earnings_date_list.append(earning_date)
        forward_div_list.append(forward_dividend_number)
        forward_yeild_list.append(forward_yeild_number)
        ex_div_date_list.append(ex_dividend_date)
        one_year_target_est_list.append(one_year_target_number)

```
