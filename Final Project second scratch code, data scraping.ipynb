```python
data = pd.read_csv("sp500.csv") 
symbol_list = list(data.Symbol)
stock_dictionary = {}
symbols_to_delete_list = []
for stock_name in stock_name_list:
    url = "https://web.archive.org/web/20170104214153/https://finance.yahoo.com/quote/"+stock_name
    request = requests.get(url)
    soup4 = BeautifulSoup(request.text, 'html.parser')
    
    if soup4.find_all("table") != []:
        # scraping for the previous close
        prev_close = str(soup4.find_all("table")[1].find_all("td")[1])\
        .split("<")[1].split(">")[1]
        try:
            prev_close = float(prev_close)
        except ValueError:
            prev_close = "N/A"


        # scraping the open value
        open_str = str(soup4.find_all("table")[1]\
                   .find_all("td")[3]).split("<")[1].split(">")[1]
        try:
            open_value = float(open_str)
        except ValueError:
            open_value = "N/A"

        # scraping for the bid, it will be a string    
        bid_str = str(soup4.find_all("table")[1]\
                      .find_all("td")[5]).split("<")[1].split(">")[1]


        # scraping for the ask, it will be a string
        ask_str = str(soup4.find_all("table")[1]\
                      .find_all("td")[7]).split("<")[1].split(">")[1]

        # scraping for the day's range, it will be a string
        day_range_str = str(soup4.find_all("table")[1]\
                            .find_all("td")[9]).split("<")[1].split(">")[1]

        #scraping for the 52 week range, it will be a string
        fifty_two_week_range_str = str(soup4.find_all("table")[1]\
                        .find_all("td")[11]).split("<")[1].split(">")[1]
        # scraping for the volume
        volume_str = str(soup4.find_all("table")[1].find_all("td")[13])\
        .split("<")[1].split(">")[1].replace(",","")

        try:
            volume_value = float(volume_str)
        except ValueError:
            volume_value = "N/A"

        # scraping for the average volume 
        avg_volume_str = str(soup4.find_all("table")[1].find_all("td")[15])\
        .split("<")[1].split(">")[1].replace(",","")

        try:
            avg_volume_value = float(avg_volume_str)
        except ValueError:
            avg_volume_value = "N/A"


        # scraping for the market cap value
        market_cap_value_str = str(soup4.find_all("table")[2].find_all("td")[1])\
                             .split("<")[1].split(">")[1].replace("B","")
        try:
            market_cap_value = float(market_cap_value_str)
        except ValueError:
            market_cap_value = "N/A"


        # scraping for the beta value
        beta_str = str(soup4.find_all("table")[2].find_all("td")[3])\
                             .split("<")[1].split(">")[1]
        try:
            beta_value = float(beta_str)
        except ValueError:
            beta_value = "N/A"

        # scraping for the PE ratio (TTM)
        pe_ratio_str = str(soup4.find_all("table")[2].find_all("td")[5])\
                             .split("<")[1].split(">")[1]
        try:
            pe_ratio_value = float(pe_ratio_str)
        except ValueError:
            pe_ratio_value = "N/A"

        #scraping for the EPS (TTM)
        eps_str = str(soup4.find_all("table")[2].find_all("td")[7])\
                             .split("<")[1].split(">")[1]
        try:
            eps_value = float(eps_str)
        except ValueError:
            eps_value = "N/A"


        # scraping for the earnings date
        earnings_date_str = str(soup4.find_all("table")[2].find_all("td")[9])\
                             .split("<")[1].split(">")[1]

        # scraping for the dividend
        try:
            dividend_value = float(str(soup4.find_all("table")[2].find_all("td")[11])\
                             .split("<")[1].split(">")[1].split(" ")[0])
        except ValueError:
            dividend_value = "N/A"

        # scraping for the yield
        if str(soup4.find_all("table")[2].find_all("td")[11])\
                             .split("<")[1].split(">")[1]\
        != "N/A":
            yield_str = str(soup4.find_all("table")[2].find_all("td")[11])\
                                     .split("<")[1].split(">")[1]\
                                     .split("(")[1].replace("%)","")
            try:
                yield_value = float(yield_str)
            except ValueError:
                yield_value = "N/A"

        else:
            yield_value = "N/A"

        # scraping for the ex-dividend date
        ex_div_date = str(soup4.find_all("table")[2].find_all("td")[13])\
                             .split("<")[1].split(">")[1]

        #scraping for the 1 year target estimate
        one_year_target_date_str = str(soup4.find_all("table")[2].find_all("td")[15])\
                             .split("<")[1].split(">")[1]

        try:
            one_year_target_date_value = float(one_year_target_date_str)
        except ValueError:
            one_year_target_date_value = "N/A"

        stock_dictionary.update({stock_name:{
            "stock_name":stock_name,
            "previous_close":prev_close,
            "open":open_value,
            "bid":bid_str,
            "ask":ask_str,
            "day_range":day_range_str,
            "52week_range": fifty_two_week_range_str,
            "volume": volume_value,
            "avg_volume":avg_volume_value,
            "market cap": market_cap_value,
            "beta": beta_value,
            "pe_ratio": pe_ratio_value,
            "eps":eps_value,
            "earnings_date":earnings_date_str,
            "forward_div":dividend_value,
            "forward_yeild": yield_value,
            "ex_div_date":ex_div_date,
            "1year_target_est":one_year_target_date_value

        }})
    elif soup4.find_all("table") == []:
        symbols_to_delete_list.append(stock_name)
```
